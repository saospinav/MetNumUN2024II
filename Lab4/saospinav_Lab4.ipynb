# Number of operations to compute a polynomial
## Extended
$$ a_5 x^5 + a_4 x^4 + a_3 x^3 + a_2 x^2 + a_1 x + a_0 = a_5*x*x*x*x*x + a_4*x*x*x*x  + a_3*x*x*x + a_2*x*x + a_1*x + a_0$$

$$ 5+4+3+2+1 = \frac{5*6}{2} (*) + 5 (+)  =20 \text{ ops} $$

$$ a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} + \cdots + a_2 x^2 + a_1 x + a_0$$

$$ \frac{n(n+1)}{2} (*) + n (+)  =\frac{n^2}{2} + \frac{3}{2} n\text{ ops} $$



## Monomials product

$$(x-r_1)(x-r_2)(x-r_3)(x-r_4)(x-r_5)$$


$$ 5 (-) + 4 (*) = 9  \text{ ops} $$

$$(x-r_1)(x-r_2)(x-r_3)\cdots (x-r_{n-1})(x-r_n)$$

$$ n-1 (*) + n (-)  =  2n-1 \text{ ops} $$

$$ n (*) + n (+)  = 2n \text{ ops} $$

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

<img src="https://raw.githubusercontent.com/gjhernandezp/NM/main/SymPyLab/sympylabwolfram7.jpg" />

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```

**My own example #8**

```
from sympy import symbols, I, N, pi, expand

# Define real symbols
a1, a2, b1, b2, c1, c2 = symbols("a1 a2 b1 b2 c1 c2", real=True)

# Define complex variables
a = a1 + I*a2
b = b1 + I*b2
c = c1 + I*c2

# Print product of a, b, and c
print("a * b * c:", a*b*c)

# Print expanded form of (a*b)/c
print("Expanded (a*b)/c:", expand(a*b/c))

# Print expanded form of a*b/c
print("Expanded a*b/c:", expand((a*b)/c))

# Print expanded form of a/(b*c)
print("Expanded a/(b*c):", expand(a/(b*c)))

# Define complex variable w
w = 1/(2*pi + I)

# Print original and N(w)
print('w =', w)
w = N(w)
print('N(w) =', w)
```

```
from sympy import symbols, exp, sin, integrate

# Define the variable
x = symbols('x')

# Define the more elaborate function
g_x = exp(-x) * sin(x)

# Integrate the function with respect to x
integral_result = integrate(g_x, x)

# Print the original function and its integral
print("Original function g(x):", g_x)
print("Integral of g(x):", integral_result)
```
